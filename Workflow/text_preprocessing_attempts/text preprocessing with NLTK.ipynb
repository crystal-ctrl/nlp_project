## Text-Preprocessing
### Different types of analysis require different data formats:
1. Corpus
2. Document-Term Matrix 

**Common data cleaning steps on all text:**
- Make text all lower case
- Remove punctuation
- Remove numerical values
- Remove common non-sensical text(/n)
- Tokenize text
- Remove stop words


**Methods:**
1. NLTK
2. spaCy

```python
import pandas as pd
import numpy as np
import pickle
import re
import string
import nltk
```

```python
with open("./data/clean_df.pkl", "rb") as read_file:
    df = pickle.load(read_file)
df.head()
```

```python
corpus = df[["cord_uid", "title", "abstract"]]
pd.set_option("max_colwidth", 150)
corpus.head()
```

```python
def clean_text(text):
    """
    Make text lowercase, remove text in square brackets, remove punctuations
    """
    text = text.lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\w*\d\w*', '', text)
    text = re.sub('[''""...]', '', text)
    text = re.sub('\n', '', text)
    return text
```

```python
# Corpus
data_clean = pd.DataFrame(corpus.abstract.apply(lambda x: clean_text(x)))
data_clean.sample(10)
```

```python
pd.set_option("max_colwidth", 250)
data_clean.sample(10)
```

```python
newStopwords = ['la', 'en', 'el', 'que', 'los', 'ci','de', 'para', 'se', 'et', 'les', 'un', 
                'las', 'por', 'con', 'des']

stopwords = nltk.corpus.stopwords.words('english')
stopwords.extend(newStopwords)
```

```python
# Document-term matrix using CountVectorizer
# try on small subset of data
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(stop_words=stopwords)
data_cv = cv.fit_transform(data_clean.abstract)
# data_dtm = pd.DataFrame(data_cv.toarray(), columns=cv.get_feature_names())
# data_dtm.index= data_clean.index
```

## Topic Modeling

```python
from sklearn.decomposition import TruncatedSVD
from sklearn.decomposition import NMF
from sklearn.metrics.pairwise import cosine_similarity
```

### LSA

```python
#LSA model
lsa = TruncatedSVD(10)
doc_topic = lsa.fit_transform(data_cv)
lsa.explained_variance_ratio_
```

```python
topic_word = pd.DataFrame(lsa.components_.round(3),
             index = list(range(1,11)),
             columns = cv.get_feature_names())
topic_word
```

```python
def display_topics(model, feature_names, no_top_words, topic_names=None):
    for ix, topic in enumerate(model.components_):
        if not topic_names or not topic_names[ix]:
            print("\nTopic ", ix)
        else:
            print("\nTopic: '",topic_names[ix],"'")
        print(", ".join([feature_names[i]
                        for i in topic.argsort()[:-no_top_words - 1:-1]]))
```

```python
display_topics(lsa, cv.get_feature_names(), 10)
```

### LDA

```python
from gensim import corpora, models, similarities, matutils

# logging for gensim (set to INFO)
import logging
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
```

```python
# Convert sparse matrix of counts to a gensim corpus
corpus = matutils.Sparse2Corpus(data_cv)
```

```python
id2word = dict((v, k) for k, v in cv.vocabulary_.items())
len(id2word)
```

```python
# Create lda model (equivalent to "fit" in sklearn)
lda = models.LdaModel(corpus=corpus, num_topics=10, id2word=id2word, passes=10)
```
