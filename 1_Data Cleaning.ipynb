## Goal:
Get the data in a clean, standard format for further analysis

## Data Cleaning
- Missing values
- Duplicates
- Remove unnecessary columns

```python
import pandas as pd
import numpy as np
import datetime
import pickle
```

```python
df = pd.read_csv("./data/metadata.csv", parse_dates=['publish_time'])
df.head()
```

```python
df.info()
```

```python
papers = df.drop(["sha", "pmcid", "pubmed_id", "license", "mag_id", "who_covidence_id",
                 "arxiv_id", "pdf_json_files", "pmc_json_files", "s2_id"], axis=1)
papers.head()
```

```python
# dropna for publish time
papers.dropna(subset=["publish_time"], inplace=True)
# create column year
papers['publish_year'] = pd.DatetimeIndex(papers.publish_time).year
```

```python
# Drop missing values
print("Before dropna: ", papers.shape)
papers.dropna(subset=["abstract"], inplace=True)
papers = papers[papers.abstract != "Unknown"]
print("After dropna: ", papers.shape)
```

```python
# Drop duplicates
papers = papers.sort_values("publish_time").reset_index(drop=True)
papers.drop_duplicates(subset=["title","abstract"],keep="last", inplace=True)
print("After drop duplicates: ", papers.shape)
```

```python
papers.drop_duplicates(subset=["cord_uid"],keep="last", inplace=True)
papers.drop_duplicates(subset=["title"],keep="last", inplace=True)
print("After drop duplicates: ", papers.shape)
```

```python
papers.head()
```

```python
papers_2020 = papers[(papers.publish_year >= 2020)].reset_index(drop=True)
```

```python
# Pickle it for later use
papers_2020.to_pickle("./data/clean_df.pkl")
```
