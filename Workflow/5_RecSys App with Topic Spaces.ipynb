```python
import pandas as pd
import numpy as np
import pickle
```

```python
df_dominant_topic = pd.read_csv("df_dominant_topic.csv")
df_dominant_topic.head()
```

```python
df_dominant_topic.dropna(inplace=True)
```

```python
df_dominant_topic.shape
```

```python
# Group top keywords for each topic
topics_sorted_df = pd.DataFrame()

topics_outdf_grpd = df_dominant_topic.groupby('Dominant_Topic')

for i, grp in topics_outdf_grpd:
    topics_sorted_df = pd.concat([topics_sorted_df, 
                                             grp.sort_values(['Topic_Perc_Contrib'], ascending=[0]).head(1)], 
                                            axis=0)

# Reset Index    
topics_sorted_df.reset_index(drop=True, inplace=True)

# Format
topics_sorted_df.columns = ['Document_No', 'Topic_Num', "Topic_Perc_Contrib", "Keywords", "Text", "Title", "url"]

# Show
topics_sorted_df = topics_sorted_df[['Topic_Num','Keywords']]
```

```python
pd.set_option("max_colwidth", 250)
topics_sorted_df
```

```python
# Group top 20 sentences under each topic
sent_topics_sorted_df = pd.DataFrame()

sent_topics_outdf_grpd = df_dominant_topic.groupby('Dominant_Topic')

for i, grp in sent_topics_outdf_grpd:
    sent_topics_sorted_df = pd.concat([sent_topics_sorted_df, 
                                             grp.sort_values(['Topic_Perc_Contrib'], ascending=[0]).head(20)], 
                                            axis=0)

# Reset Index    
sent_topics_sorted_df.reset_index(drop=True, inplace=True)

# Format
sent_topics_sorted_df.columns = ['Document_No', 'Topic_Num', "Topic_Perc_Contrib", "Keywords", "Text", "Title", "url"]

# Show
sent_topics_sorted_df.head(10)
```

```python
sent_topics_sorted_df.to_csv("./app/app_ready.csv", index=False)
```

```python
def topic_relative_paper(topic_num):
    df = sent_topics_sorted_df[sent_topics_sorted_df['Topic_Num']== topic_num].reset_index(drop=True)
    for i in range(20):
        print(df.Title),'\n')
```

```python
topic_relative_paper(27)
```
